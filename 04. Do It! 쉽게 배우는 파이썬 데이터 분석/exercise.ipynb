## 분석 도전  
midwest.csv는 미국 동북중부(East North Central States) 437개 지역의 인구통계 정보를 담고 있습니다. midwest.csv를 이용해 데이터 분석 문제를 해결해 보세요.

```python
# 1. midwest.csv 를 불러와 데이터의 특징을 파악하세요.
import pandas as pd
import numpy as np

midwest = pd.read_csv('./Data/midwest.csv')
midwest.head()
```

```python
midwest.info()
```

```python
# 2. poptotal(전체인구) 변수를 total로, popasian(아시아 인구) 변수를 asian으로 수정하세요.
midwest_copy = midwest.copy()
midwest_copy.rename(columns={'poptotal': 'total', 'popasian': 'asian'}, inplace=True)
midwest_copy.head()
```

```python
midwest_copy.info()
```

```python
midwest_copy.loc[:,['PID','total','asian']]
```

```python
# 3. total, asian 변수를 이용해 '전체 인구 대비 아시아 인구 백분율' 파생변수를 추가하고,
# 히스토그램을 만들어 분포를 살펴보세요.
midwest_copy['asian-per-total'] = midwest_copy['asian'] / midwest_copy['total'] * 100
midwest_copy.loc[0:5,['PID','total','asian', 'asian-per-total']]
```

```python
midwest_copy['asian-per-total'].plot.hist()
```

```python
# 4. 아시아 인구 백분율 전체 평균을 구하고, 평균을 초과하면 'large', 그 외에는 'small'을
# 부여한 파생변수를 만들어 보세요.
percent = midwest_copy['asian-per-total'].mean()
```

```python
midwest_copy['per-mean'] = np.where(midwest_copy['asian-per-total'] > percent, 'large', 'small')
midwest_copy.loc[0:5,['PID','total','asian', 'asian-per-total', 'per-mean']]
```

```python
# 5. 'large'와 'small'에 해당하는 지역이 얼마나 많은지 빈도표와 빈도 막대 그래프를
# 만들어 확인해 보세요.
midwest_copy['per-mean'].value_counts()
```

```python
asian_per_count = midwest_copy['per-mean'].value_counts()
asian_per_count.plot.bar(rot=0)
```

```python
# 2. 극단치 기준값 구하기
# (1) 1사분위수, 3사분위수 구하기
pct25 = mpg['hwy'].quantile(.25)
pct25
```

```python
pct75 = mpg['hwy'].quantile(.75)
pct75
```

```python
# (2) IQR(inter quantile range) 구하기 : 1사분위수와 3사분위수의 거리
iqr = pct75 - pct25
iqr
```

```python
# (3) 하한, 상한 구하기
# - 하한 : 1사분위수보다 'IQR의 1.5배'만큼 더 작은 값
min = pct25 - 1.5*iqr
print(min)
# - 상한 : 3사분위수보다 'IQR의 1.5배'만큼 더 큰 값
max = pct25 + 1.5*iqr
print(max)
```
