```python
import folium

# 강남역 근처 좌표 (위도, 경도)
gangnam_coords = [37.49795, 127.02763] 
    
# Folium 지도 객체 생성
# 중심 좌표와 초기 확대/축소 레벨(zoom_start=18) 설정
seoul_map = folium.Map(
    location=gangnam_coords,
    zoom_start=18,
    tiles="OpenStreetMap"
)
    
# 지도 중심에 마커 추가
# popup에 마커에 대한 설명 추가
folium.Marker(
    location=gangnam_coords,
    popup="강남역",
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(seoul_map)


seoul_map
```

```python
# 지도를 HTML 파일로 저장
seoul_map.save("gangnam_map.html")

print("지도가 'gangnam_map.html' 파일로 성공적으로 저장되었습니다.")
```
